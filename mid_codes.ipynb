```python
# DDA
# Calculate the points between the starting point (1,1) and the ending point (4,3).

import matplotlib.pyplot as plt

# DDA Line Drawing Function with integer rounding
def dda(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1

    for _ in range(steps + 1):
        # Round to nearest integer and format as 2 decimal places
        px = round(x)
        py = round(y)
        points.append((float(f"{px:.2f}"), float(f"{py:.2f}")))
        x += x_inc
        y += y_inc
    return points

# Starting and ending points
x1, y1 = 1, 1
x2, y2 = 4, 3

# Get points
points = dda(x1, y1, x2, y2)
print("Calculated Points (Rounded to Integer):")
for p in points:
    print(p)

# Plotting
plt.figure(figsize=(6,6))
x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

# Draw line
plt.plot(x_vals, y_vals, marker='o', color='blue')

# Mark points with labels
for i, (x, y) in enumerate(points):
    label = f"P{i}({x},{y})"
    plt.text(x + 0.05, y + 0.05, label, fontsize=9)

# Formatting
plt.title("DDA Line Drawing (Rounded to Integer Points)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
# Bresenham's Line Drawing Algorithm
# Calculate the points between the starting point (9,18) and the ending point (14,22).

import matplotlib.pyplot as plt

# Bresenham's Line Drawing Algorithm
def bresenham(x1, y1, x2, y2):
    points = []
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    err = dx - dy

    while True:
        points.append((float(f"{x1:.2f}"), float(f"{y1:.2f}")))  # store as float for formatting
        if x1 == x2 and y1 == y2:
            break
        e2 = 2 * err
        if e2 > -dy:
            err -= dy
            x1 += sx
        if e2 < dx:
            err += dx
            y1 += sy
    return points

# Starting and ending points
x1, y1 = 9, 18
x2, y2 = 14, 22

# Get points
points = bresenham(x1, y1, x2, y2)
print("Calculated Points (Bresenham):")
for p in points:
    print(p)

# Plotting
plt.figure(figsize=(6,6))
x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

# Draw line
plt.plot(x_vals, y_vals, marker='o', color='green')

# Mark points with labels
for i, (x, y) in enumerate(points):
    label = f"P{i}({x},{y})"
    plt.text(x + 0.05, y + 0.05, label, fontsize=9)

# Formatting
plt.title("Bresenham's Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
# Mid-Point Line Drawing Algorithm
# Calculate the points between the starting point (5,12) and the ending point (16,22).

import matplotlib.pyplot as plt

# Mid-Point Line Drawing Algorithm
def midpoint_line(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1
    d = dy - (dx / 2)   # Initial decision parameter
    x, y = x1, y1

    points.append((float(f"{x:.2f}"), float(f"{y:.2f}")))

    while x < x2:
        x += 1
        if d < 0:
            d = d + dy
        else:
            d = d + (dy - dx)
            y += 1
        points.append((float(f"{x:.2f}"), float(f"{y:.2f}")))
    return points

# Starting and ending points
x1, y1 = 5, 12
x2, y2 = 16, 22

# Get points
points = midpoint_line(x1, y1, x2, y2)
print("Calculated Points (Mid-Point Line Drawing):")
for p in points:
    print(p)

# Plotting
plt.figure(figsize=(6,6))
x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

# Draw line
plt.plot(x_vals, y_vals, marker='o', color='red')

# Mark points with labels
for i, (x, y) in enumerate(points):
    label = f"P{i}({x},{y})"
    plt.text(x + 0.05, y + 0.05, label, fontsize=9)

# Formatting
plt.title("Mid-Point Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
# Bresenham's Circle Drawing Algorithm
# Given the center point coordinates (0,0) and the radius of 8, generate all the points to form the circle.

import matplotlib.pyplot as plt

# Bresenham's Circle Drawing Algorithm
def bresenham_circle(xc, yc, r):
    points = []
    x = 0
    y = r
    d = 3 - 2 * r

    def plot_circle_points(xc, yc, x, y):
        pts = [
            (xc + x, yc + y),
            (xc - x, yc + y),
            (xc + x, yc - y),
            (xc - x, yc - y),
            (xc + y, yc + x),
            (xc - y, yc + x),
            (xc + y, yc - x),
            (xc - y, yc - x),
        ]
        for px, py in pts:
            points.append((float(f"{px:.2f}"), float(f"{py:.2f}")))

    while x <= y:
        plot_circle_points(xc, yc, x, y)
        if d < 0:
            d += 4 * x + 6
        else:
            d += 4 * (x - y) + 10
            y -= 1
        x += 1

    # Remove duplicates
    unique_points = []
    for p in points:
        if p not in unique_points:
            unique_points.append(p)
    return unique_points

# Center and radius
xc, yc = 0, 0
r = 8

# Get points
points = bresenham_circle(xc, yc, r)
print("Calculated Points (Bresenham's Circle Drawing):")
for p in points:
    print(p)

# Sort points by angle for proper connection
import math
points_sorted = sorted(points, key=lambda p: math.atan2(p[1] - yc, p[0] - xc))

# Plotting
plt.figure(figsize=(6,6))
x_vals = [p[0] for p in points_sorted]
y_vals = [p[1] for p in points_sorted]

# Close the loop by adding the first point at the end
x_vals.append(x_vals[0])
y_vals.append(y_vals[0])

# Draw connected circle
plt.plot(x_vals, y_vals, marker='o', color='purple')

# Mark points with labels
for i, (x, y) in enumerate(points_sorted):
    label = f"P{i}({x},{y})"
    plt.text(x + 0.2, y + 0.2, label, fontsize=8)

# Formatting
plt.title("Bresenham's Circle Drawing Algorithm (Connected)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
# Mid-Point Circle Drawing Algorithm
# Given the center point coordinates (4,4) and the radius of 10, generate all the points to form the circle.


import matplotlib.pyplot as plt
import math

# Mid-Point Circle Drawing Algorithm
def midpoint_circle(xc, yc, r):
    points = []
    x = 0
    y = r
    p = 1 - r  # Initial decision parameter

    def plot_circle_points(xc, yc, x, y):
        pts = [
            (xc + x, yc + y),
            (xc - x, yc + y),
            (xc + x, yc - y),
            (xc - x, yc - y),
            (xc + y, yc + x),
            (xc - y, yc + x),
            (xc + y, yc - x),
            (xc - y, yc - x),
        ]
        for px, py in pts:
            if (px, py) not in points:
                points.append((float(f"{px:.2f}"), float(f"{py:.2f}")))

    while x <= y:
        plot_circle_points(xc, yc, x, y)
        if p < 0:
            p += 2 * x + 3
        else:
            p += 2 * (x - y) + 5
            y -= 1
        x += 1

    return points

# Center and radius
xc, yc = 4, 4
r = 10

# Get points
points = midpoint_circle(xc, yc, r)
print("Calculated Points (Mid-Point Circle Drawing):")
for p in points:
    print(p)

# Sort points by angle for proper connection
points_sorted = sorted(points, key=lambda p: math.atan2(p[1] - yc, p[0] - xc))

# Plotting
plt.figure(figsize=(7,7))
x_vals = [p[0] for p in points_sorted]
y_vals = [p[1] for p in points_sorted]

# Close loop
x_vals.append(x_vals[0])
y_vals.append(y_vals[0])

# Draw connected circle
plt.plot(x_vals, y_vals, marker='o', color='blue')

# Mark points with labels
for i, (x, y) in enumerate(points_sorted):
    label = f"P{i}({x},{y})"
    plt.text(x + 0.2, y + 0.2, label, fontsize=8)

# Formatting
plt.title("Mid-Point Circle Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
